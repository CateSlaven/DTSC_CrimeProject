```python
# Identify outliers using IQR method


for column in crimes_df.columns:
    Q1 = crimes_df[column].quantile(0.25)
    Q3 = crimes_df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = crimes_df[(crimes_df[column] < lower_bound) | (crimes_df[column] > upper_bound)]
    
    # Print results
    print(f"Column: {column}")
    print(f"Q1: {Q1}")
    print(f"Q3: {Q3}")
    print(f"IQR: {IQR}")
    print(f"Outliers:\n{outliers}\n")

# for col in crimes_df.columns:
#     plt.figure(figsize=(8, 6))
#     sns.boxplot(x=crimes_df[col])
#     plt.title(col)
#     plt.show()

#     # Set up the figure and axes
# plt.figure(figsize=(12, 8))
# sns.boxplot(data=crimes_df)

# # Add titles and labels
# plt.title('Boxplots of Violent Crime Metrics', fontsize=16)
# plt.xlabel('Crime Metrics', fontsize=14)
# plt.ylabel('Values', fontsize=14)

# # Show the plot
# plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
# plt.show()

crimes_df
```
